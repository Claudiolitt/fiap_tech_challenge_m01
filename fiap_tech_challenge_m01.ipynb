```python
import pandas as pd
from fastapi import FastAPI, HTTPException, Response
import requests
import io

app = FastAPI()

# Função para buscar dados da Embrapa
def fetch_data_from_embrapa(url):
    try:
        response = requests.get(url)
        response.raise_for_status()
        return response.content.decode('utf-8')  # Retorna o conteúdo do CSV como string
    except requests.exceptions.RequestException as e:
        raise HTTPException(status_code=500, detail=str(e))

@app.get("/producao")
def get_producao():
    url = "http://vitibrasil.cnpuv.embrapa.br/download/Producao.csv"
    csv_data = fetch_data_from_embrapa(url)

    # Ler o CSV diretamente usando pandas e io.StringIO
    df = pd.read_csv(io.StringIO(csv_data), delimiter=';')  # Ajuste o delimitador conforme necessário

    # Ajustar a configuração para exibir todas as colunas e mais linhas
    pd.set_option('display.max_columns', None)
    pd.set_option('display.max_rows', None)

    # Exibir o DataFrame completo
    #print(df.to_string())
    return df

@app.get("/processamento")
def get_processamento():
    url = "http://vitibrasil.cnpuv.embrapa.br/download/ProcessaViniferas.csv"
    csv_data = fetch_data_from_embrapa(url)

    # Ler o CSV diretamente usando pandas e io.StringIO
    df = pd.read_csv(io.StringIO(csv_data), delimiter=';')  # Ajuste o delimitador conforme necessário

    # Ajustar a configuração para exibir todas as colunas e mais linhas
    pd.set_option('display.max_columns', None)
    pd.set_option('display.max_rows', None)

    # Exibir o DataFrame completo
    #print(df.to_string())
    return df




@app.get("/comercializacao")
def get_comercializacao():
    url = "http://vitibrasil.cnpuv.embrapa.br/download/Comercio.csv"
    csv_data = fetch_data_from_embrapa(url)

    # Ler o CSV diretamente usando pandas e io.StringIO
    df = pd.read_csv(io.StringIO(csv_data), delimiter=';')  # Ajuste o delimitador conforme necessário

    # Ajustar a configuração para exibir todas as colunas e mais linhas
    pd.set_option('display.max_columns', None)
    pd.set_option('display.max_rows', None)

    # Exibir o DataFrame completo
    #print(df.to_string())
    return df

@app.get("/importacao")
def get_importacao():
    url = "http://vitibrasil.cnpuv.embrapa.br/download/ImpVinhos.csv"
    csv_data = fetch_data_from_embrapa(url)

    # Ler o CSV diretamente usando pandas e io.StringIO
    df = pd.read_csv(io.StringIO(csv_data), delimiter=';')  # Ajuste o delimitador conforme necessário

    # Ajustar a configuração para exibir todas as colunas e mais linhas
    pd.set_option('display.max_columns', None)
    pd.set_option('display.max_rows', None)

    # Exibir o DataFrame completo
    #print(df.to_string())
    return df

@app.get("/exportacao")
def get_exportacao():
    url = "http://vitibrasil.cnpuv.embrapa.br/download/ExpVinho.csv"
    csv_data = fetch_data_from_embrapa(url)

    # Ler o CSV diretamente usando pandas e io.StringIO
    df = pd.read_csv(io.StringIO(csv_data), delimiter=';')  # Ajuste o delimitador conforme necessário

    # Ajustar a configuração para exibir todas as colunas e mais linhas
    pd.set_option('display.max_columns', None)
    pd.set_option('display.max_rows', None)

    # Exibir o DataFrame completo
    #print(df.to_string())
    return df
```
